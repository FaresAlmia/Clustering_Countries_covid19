```python
import pandas as pd
```

```python
data_url=r'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
```

```python
confirmed_df=pd.read_csv(data_url)
```

```python
df=confirmed_df.groupby(['Country/Region']).sum()
```

```python
df.drop(columns=['Lat','Long'], inplace=True)
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=10)
km.fit(df)
labels=km.predict(df)
```

```python
from sklearn.mixture import GaussianMixture
gmm=GaussianMixture(n_components=10,covariance_type='diag')
gmm.fit(df)
labels=gmm.predict(df)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(df)
Xn=ss.transform(df)
```

```python
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(Xn)
Xp=pca.transform(Xn)
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(Xp[labels==0,0],Xp[labels==0,1],'go',label='cluster 0')
# plt.plot(Xp[labels==1,0],Xp[labels==1,1],'ro',label='cluster 1')
plt.plot(Xp[labels==2,0],Xp[labels==2,1],'bo',label='cluster 2')
plt.plot(Xp[labels==3,0],Xp[labels==3,1],'mo',label='cluster 3')
plt.plot(Xp[labels==4,0],Xp[labels==4,1],'ko',label='cluster 4')
plt.plot(Xp[labels==5,0],Xp[labels==5,1],'yo',label='cluster 5')
plt.plot(Xp[labels==6,0],Xp[labels==6,1],'r*',label='cluster 6')
plt.plot(Xp[labels==7,0],Xp[labels==7,1],'b*',label='cluster 7')
# plt.plot(Xp[labels==8,0],Xp[labels==8,1],'g*',label='cluster 8')
plt.legend()
```
